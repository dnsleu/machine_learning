# Data preparation

### Clean, transform and prepare a reproducible model table. Saved in `data/processed`

## 1. Libraries & settings

```python
import sys
import pandas as pd
from pathlib import Path
```

## 2. Paths & load data

```python
PROJECT_DIR = Path.cwd().resolve().parents[0]
sys.path.append(str(PROJECT_DIR))

from src.preprocessing import (
    PrepConfig, clean_columns_basic, apply_column_renames, build_model_table, save_processed
)

DATA_RAW  = PROJECT_DIR / "data" / "raw"
DATA_PROCESSED = PROJECT_DIR / "data" / "processed"
DATA_FILE = next(DATA_RAW.glob("*.csv"))
df_raw = pd.read_csv(DATA_FILE)
df_raw.head()
```

## 3. Column config

```python
cfg = PrepConfig(
    feature_cols=["Age", "Annual_income_thousands", "Spending_score"],
    category_cols=["Gender"],
    rename_map={
        "Spending Score (1-100)": "Spending_score",
        "Annual Income (k$)": "Annual_income_thousands"
    }
)
```

## 4. Data cleaning, renaming

```python
df = apply_column_renames(df_raw, cfg.rename_map)
df = clean_columns_basic(df)
df
```

## 5. Feature columns

```python
model_df = build_model_table(df, cfg)
model_df
```

## 7. Save processed dataset

```python
out_csv = DATA_PROCESSED / "model_table.csv"
save_processed(model_df, out_csv)
```
